```python
import matplotlib.pyplot as plt
import numpy as np

species = ("Whisper-tiny.en", "wav2vec2.0")
penguin_means = {
    'Inference': (.43,.55),
    'Fine-tuning (Whisper) / self-supervised learning (wav2vec2.0)': (0.08, 0.29),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Word error rate')
ax.set_title('WER for Whisper and wav2vec2.0 regimes')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=2, prop={'size': 6})
ax.set_ylim(0, 1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

species = ("Whisper-tiny.en", "wav2vec2.0")
penguin_means = {
    'Inference': (.43, .55),
    'Fine-tuning (Whisper) / self-supervised learning (wav2vec2.0)': (0.08, 0.29),
}

x = np.arange(len(species))  # the label locations
width = 0.35  # the width of the bars
multiplier = 0

fig, ax = plt.subplots()

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Set bar colors
rects[0].set_color('red')
rects[1].set_color('red')



# Add some text for labels, title, and custom x-axis tick labels
ax.set_ylabel('Word error rate')
ax.set_title('WER for Whisper and wav2vec2.0 regimes')
ax.set_xticks(x + width / 2, species)
ax.legend(loc='upper left', ncols=2, prop={'size': 6})
ax.set_ylim(0, 1)

plt.figure(figsize=(12,6))
plt.figure().set_figwidth(100)

plt.show()
```
